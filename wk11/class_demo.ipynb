Here we use text data from rdr2 review to do the demonstration.
RDR2 (Red Dead Redemption 2) is a game produced by Rockstar Studion. It's a story about Arthur, a western gangster, who strived to survive in the harsh condition.
Therefore, I expect to see words like Arthur(Person), Rockstar(Organization), Blackwater/Valentine(The place where the story goes) popping out and being identified by NER.

Text Link: https://www.ign.com/articles/2018/10/25/red-dead-redemption-2-review

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import TreebankWordTokenizer
from nltk.tag import pos_tag
from nltk import ne_chunk, Tree
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
f = open('rdr2.txt', 'r')
ex = f.readlines()
f.close()
len(ex)
```

```python
strEx = ' '.join(str(x) for x in ex)
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
sent = preprocess(strEx)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
tree1 = nltk.Tree('NP', ['Alick'])
print(tree1)
tree2 = nltk.Tree('N', ['Alick', 'Rabbit'])
print(tree2)
tree3 = nltk.Tree('S', [tree1, tree2])
print(tree3.label())
```

```python
NER = ne_chunk(nltk.pos_tag(word_tokenize(strEx)), binary = False)
```

```python
type(NER)
```

```python
print(NER)
```

```python
def get_continuous_chunks(text, label):
    chunked = ne_chunk(pos_tag(word_tokenize(text)))
    prev = None
    continuous_chunk = []
    current_chunk = []
    
    for subtree in chunked:
        if type(subtree) == Tree and subtree.label() == label:
            current_chunk.append(" ".join([token for token, pos in subtree.leaves()]))
        elif current_chunk:
            named_entity = " ".join(current_chunk)
            if named_entity not in continuous_chunk:
                continuous_chunk.append(named_entity)
                current_chunk = []
        else:
            continue
    return continuous_chunk
```

```python
get_continuous_chunks(strEx, "GPE")
```

Here the word "Dutch" is a place/country in the real world; however, in this game it's a name of a character.

```python
get_continuous_chunks(strEx, "PERSON")
```

As we expected, we can see a lot of names like Arthur, John...But what's interesting here it's that NER tend to combine strang words as a person, like "Coffee Arthur", "Arthur Guns"...

```python
get_continuous_chunks(strEx, "ORGANIZATION")
```

There's a conflict for the word "Red Dead": sometimes it is identified as a name, but somtimes it is identified as an organization.

### Conclusion
This method can easily extract the "point" we would like to see in a long article, like who's in the story, where did it take place...etc. While in R, we use TF-IDF or other text weight to extract the importance of individual words, which in comparison, could not really tell the story.
For this example, in R we can see top 5 words which has the most frequencies are: world, dead, arthur, redemption, feel. It would be hard for us to understsand the story in this way, especially when this article contains lots of information that could possibly be mentioned once or twice.
